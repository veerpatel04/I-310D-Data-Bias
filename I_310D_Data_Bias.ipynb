My hypothesis is, when analyzing toxic comments about rich individuals, the Perspective API is more likely to make mistakes than when studying toxic comments about poor people. I have collected ten comments about rich people (five toxic and five non-toxic), ten comments about poor people (five toxic and five non-toxic), and I have tested them using the Perspective API client to see whether it accurately classifies them as toxic or non-toxic.

Load the dataset in 3 column format : socioeconomic status, self_input, comment

```python
import numpy as np
import pandas as pd
from googleapiclient import discovery
import json

df = pd.read_csv("I 310D-Data Bias.csv")
df
```

Function that uses the Perspective API client to score each comment

```python
def get_perspective_API_value(comment):
    API_KEY = "AIzaSyAupwJUXG4hJgMCDXCjlicV4s6Yp0s3y6Y"

    client = discovery.build(
        "commentanalyzer",
        "v1alpha1",
        developerKey=API_KEY,
        discoveryServiceUrl="https://commentanalyzer.googleapis.com/$discovery/rest?version=v1alpha1",
        static_discovery=False,
    )

    analyze_request = {
        "comment": {"text": comment},
        "requestedAttributes": {"TOXICITY": {}},
    }

    response = client.comments().analyze(body=analyze_request).execute()
    perspective_API_value = response['attributeScores']['TOXICITY']['spanScores'][0]['score']['value']

    return perspective_API_value
```

Use the get_perspective_API_value function to score each comment and create a new column perspective_API_value

```python
df["perspective_API_value"] = df["Comment"].apply(get_perspective_API_value)
df
```

Create a new column api_result to label each comment based on the perspective_API_value

```python
df["api_result"] = [1 if result > 0.5 else 0 for result in df["perspective_API_value"]]
df
```

Convert Self_Input to binary

```python
df["Self_Input"] = [1 if x=='Toxic' else 0 for x in df['Self_Input']]
df
```

Select relevant columns for testing

```python
df = df[["Socioeconomic Status", "Self_Input", "api_result"]]
df
```

Extract indices for Rich and Poor comments (first step of finding accuracy)

```python
type_column = df["Socioeconomic Status"]

rich_indices = []
poor_indices = []
y_actual = [y for y in df['Self_Input']]
y_predicted = [y for y in df['api_result']]

for i in range(len(type_column)):
    if type_column[i] == "Rich":
        rich_indices.append(i)
    else:
        poor_indices.append(i)

y_actual_poor = [y_actual[i] for i in poor_indices]
y_predicted_poor = [y_predicted[i] for i in poor_indices]

y_actual_rich = [y_actual[i] for i in rich_indices]
y_predicted_rich = [y_predicted[i] for i in rich_indices]

print(len(poor_indices))
print(len(rich_indices))
```

Define function to find class wise accuracy of comments

```python
def class_wise_acc(y_actual, y_predicted):
    total_p = 0
    total_n = 0
    TP=0
    TN=0
    for i in range(len(y_predicted)):
        if y_actual[i]==1:
            total_p = total_p+1
            if y_actual[i]==y_predicted[i]:
               TP=TP+1
        if y_actual[i]==0:
            total_n=total_n+1
            if y_actual[i]==y_predicted[i]:
               TN=TN+1
    return(TP/total_p, TN/total_n)

acc_tox_poor, acc_ntox_poor = class_wise_acc(y_actual_poor, y_predicted_poor)
acc_tox_rich, acc_ntox_rich = class_wise_acc(y_actual_rich, y_predicted_rich)

print (f"Poor and Non-Toxic detection accuracy = {acc_ntox_poor}")
print (f"Poor and Toxic detection accuracy = {acc_tox_poor}")
print (f"Rich and Non-Toxic detection accuracy = {acc_ntox_rich}")
print (f"Rich and Toxic detection accuracy = {acc_tox_rich}")
```
